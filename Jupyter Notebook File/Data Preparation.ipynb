```python
base_path = r"C:\Users\wmkoa\Downloads\Fund data\AIA"

dir_lists = os.listdir(base_path)
df_list_all = []
for i in range(1,len(dir_lists)+1):
    dir_link = os.path.join(base_path, str(i))
    dir_lists = os.listdir(dir_link)
    df_list_fund = []
    for filename in dir_lists:
        file_path = os.path.join(dir_link, filename)
        df = pd.read_csv(file_path,names=['Date', 'Price'],index_col=[0],parse_dates=[0])
        df['Price'] = df['Price'].str.replace("\t","")
        df.rename(columns={'Price': df['Price'][0] +'_Price'}, inplace=True)
        df = df[pd.to_numeric(df.iloc[:,0], errors='coerce').notnull()]
        df_list_fund.append(df)
    df_list_all.append(pd.concat(df_list_fund).sort_index())

# for item in df_list_all:
#     item.reset_index(inplace=True)
#     #item.reset_index(drop=True, inplace=True)
# #     df = pd.merge(df, item, on='Date', how='outer')
# df = pd.concat(df_list_all, axis=1)
```

```python
# Export indivdual fund price to csv
for i in range(len(df_list_all)):
    df_list_all[i].to_csv(os.path.join(base_path,(df_list_all[i].columns[0] + '.csv')), encoding='utf_8_sig')
```

```python
# Export combined fund price to csv
df = pd.concat(df_list_all, axis=1).sort_index()
df.to_csv(os.path.join(base_path,'combined_fund_prices.csv'), encoding='utf_8_sig')
```
